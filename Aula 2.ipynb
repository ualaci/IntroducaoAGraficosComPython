## Funções, bibliotecas numpy e matplotlib

Imagine que você tem uma tarefa a ser feita várias vezes, por exemplo, somar dois números. Em vez de escrever o código para essa tarefa repetidamente, você pode criar uma função que a execute e reutilizá-la sempre que precisar.

Uma função em Python é como uma caixa preta que recebe alguns valores de entrada, realiza algumas ações e pode retornar um resultado. Ela é definida usando a palavra-chave def, seguida pelo nome da função e, entre parênteses, os parâmetros que ela recebe (se houver).

Vejamos agora um exemplo de uma função simples que soma dois números:

```python
def somar(a, b):
    resultado = a + b
    return resultado

resultado = somar(3, 5)
print("resultado:", resultado)  # Saída: 8
```

Neste exemplo, chamamos a função somar passando os valores 3 e 5 como argumentos. O resultado retornado pela função é armazenado na variável resultado e, em seguida, é impresso na saída.

Uma função pode ter mais complexidade e realizar várias ações dentro dela, incluindo estruturas de controle condicional (if, else, elif) e estruturas de repetição (while, for), entre outras coisas. Como podemos ver no exemplo a seguir:

```python
def contar_vogais(palavra):
    contador = 0

    for letra in palavra:
        if letra.lower() in ['a', 'e', 'i', 'o', 'u']:
            contador += 1

    if contador == 1:
        print("A palavra", palavra, "possui 1 vogal.")
    else:
        print("A palavra", palavra, "possui", contador, "vogais.")
```

Neste exemplo, chamamos a função contar_vogais com duas palavras diferentes. Ela conta o número de vogais em cada palavra e imprime a mensagem correspondente.

Esse exemplo demonstra como uma função pode usar estruturas de controle condicional (if-else) e estruturas de repetição (for) para realizar uma tarefa específica.

Em resumo, as funções são blocos de código reutilizáveis em Python que permitem que você organize seu programa e evite repetição de código. Elas ajudam a tornar seu código mais legível, modular e fácil de manter.

## Biblioteca numpy

A biblioteca NumPy é uma das bibliotecas mais populares em Python para computação numérica. Ela fornece uma estrutura de dados chamada de array, que é semelhante a uma lista, mas com várias vantagens. Com o NumPy, você pode executar operações matemáticas de forma eficiente em grandes conjuntos de dados.
Vejamos um exemplo a seguir:

```python
import numpy as np

def calcular_media(array):
    media = np.mean(array)
    return media

valores = np.array([2, 4, 6, 8, 10])
resultado = calcular_media(valores)
print("A média dos valores é:", resultado)
```

Neste exemplo, criamos um array valores utilizando a função array da biblioteca NumPy. Em seguida, chamamos a função calcular_media passando o array como argumento. A função calcula a média dos valores do array utilizando o NumPy e retorna o resultado. Por fim, imprimimos o resultado na saída.
Em seguida podemos ver mais algumas possibilidades de operações utilizando a biblioteca numpy:

```python
def realizar_operacoes(array1, array2):
    # Soma dos arrays
    soma = array1 + array2
    print("Soma:", soma)

    # Multiplicação dos arrays
    multiplicacao = array1 * array2
    print("Multiplicação:", multiplicacao)

    # Exponenciação do array1
    exponenciacao = np.exp(array1)
    print("Exponenciação:", exponenciacao)

# Chamar a função com os arrays
array1 = np.array([1, 2, 3])
array2 = np.array([4, 5, 6])
realizar_operacoes(array1, array2)
```

Neste exemplo, a função realizar_operacoes recebe dois arrays como parâmetros. Dentro da função, as operações de soma, multiplicação e exponenciação são realizadas usando os arrays fornecidos. Os resultados são impressos na saída.

Em seguida, chamamos a função realizar_operacoes passando os arrays array1 e array2.

## Biblioteca MatplotLib
O Matplotlib é uma biblioteca amplamente utilizada em Python para criação de gráficos e visualização de dados. Ele oferece uma ampla gama de funcionalidades para criar diferentes tipos de gráficos, como gráficos de linhas, gráficos de dispersão, histogramas, gráficos de barras, gráficos de pizza, etc.

Vejamos agora um exemplo que utiliza o Matplotlib para criar um gráfico de linhas a partir de um array NumPy:

```python
import numpy as np
import matplotlib.pyplot as plt

def plotar_grafico(array):
    # Criar um array para o eixo x
    x = np.arange(len(array))

    # Plotar o gráfico de linhas
    plt.plot(x, array)

    # Configurar rótulos e título do gráfico
    plt.xlabel('Eixo x')
    plt.ylabel('Eixo y')
    plt.title('Gráfico de Linhas')

    # Exibir o gráfico
    plt.show()

# Chamar a função para plotar o gráfico
array = np.array([1, 4, 2, 6, 8, 3])
plotar_grafico(array)
```

Neste exemplo, a função plotar_grafico recebe um array como parâmetro. Dentro da função, utilizamos a função arange do NumPy para criar um array para o eixo x com base no tamanho do array fornecido. Em seguida, utilizamos a função plot do Matplotlib para criar o gráfico de linhas com os valores do array.

Configuramos os rótulos dos eixos x e y, bem como o título do gráfico, usando as funções xlabel, ylabel e title do Matplotlib.

Por fim, chamamos a função show do Matplotlib para exibir o gráfico.

Ao executar esse código, será exibido um gráfico de linhas com base nos valores do array fornecido.

Vejamos agora um outro exemplo de gráfico para um experimento de medição da posição x tempo

```python
import numpy as np
import matplotlib.pyplot as plt

def plotar_posicao_tempo(tempo, posicao):
    # Plotar o gráfico de posição em função do tempo
    plt.plot(tempo, posicao, marker='o', linestyle='-', color='b')

    # Configurar rótulos e título do gráfico
    plt.xlabel('Tempo (s)')
    plt.ylabel('Posição (m)')
    plt.title('Posição em Função do Tempo')

    # Exibir o gráfico
    plt.show()

# Dados do experimento
tempo = np.array([0, 1, 2, 3, 4, 5])  # Tempo (s)
posicao = np.array([0, 5, 10, 15, 20, 25])  # Posição (m)

# Chamar a função para plotar o gráfico de posição em função do tempo
plotar_posicao_tempo(tempo, posicao)
```

Neste exemplo, a função plotar_posicao_tempo recebe dois arrays como parâmetros: tempo para os valores de tempo e posicao para os valores de posição do objeto.

Dentro da função, utilizamos a função plot do Matplotlib para criar o gráfico de posição em função do tempo. Configuramos o estilo da linha com linestyle='-' para uma linha sólida, e marcadores em forma de círculo com marker='o'. A cor da linha é definida como azul (color='b').

Configuramos os rótulos dos eixos x e y, bem como o título do gráfico, usando as funções xlabel, ylabel e title do Matplotlib.

Por fim, chamamos a função show do Matplotlib para exibir o gráfico.
